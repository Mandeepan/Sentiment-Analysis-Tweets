```python
import pandas as pd
import json
from pandas.io.json import json_normalize
import numpy as np
```

```python
#jupyter notebook setting: to show the full texts in the output window
pd.set_option('display.max_colwidth', -1)
```

```python
data=json.load((open('data/world_bank_projects.json')))
```

```python
data_df=json_normalize(data)
```

```python
data_df.columns
```

### 1. Find the 10 countries with most projects

```python
result_one=data_df.groupby(['project_name'])['countryname'].apply(lambda x: x.value_counts().index[0]).reset_index()
result_one
```

```python
result_one=result_one.countryname.value_counts()[0:10]
result_one
```

### 2. Find the top 10 major project themes (using column 'mjtheme_namecode')

```python
major_theme=pd.DataFrame(json_normalize(data,'mjtheme_namecode'))
major_theme
```

```python
result_two=major_theme.name.value_counts()[0:10]
result_two
```
